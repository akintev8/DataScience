```python
import seaborn as sns 
import matplotlib.pyplot as plt 
import pandas as pd 
df = pd.read_csv("https://raw.githubusercontent.com/luiseduardogfranca/students-performance/master/StudentsPerformance.csv") 
```

```python
#1
sns.histplot(df["math score"], kde=True) 
plt.title("Распределение оценок по математике") 
plt.show() 
sns.histplot(df["reading score"], kde=True) 
plt.title("Распределение оценок по чтению") 
plt.show() 
sns.histplot(df["writing score"], kde=True) 
plt.title("Распределение оценок по письму") 
plt.show()
```

```python
#2 
sns.boxplot(x="gender", y="math score", data=df) 
plt.title("Сравнение результатов по математике между полами") 
plt.show() 
sns.boxplot(x="gender", y="writing score", data=df) 
plt.title("Сравнение результатов по письму между полами") 
plt.show() 
sns.boxplot(x="gender", y="reading score", data=df) 
plt.title("Сравнение результатов по чтению между полами") 
plt.show() 
```

```python
#3 
plt.figure(figsize=(10,5)) 
sns.barplot(x="parental level of education", y="math score", data=df) 
plt.title("Средний балл по математике в зависимости от образования родителей") 
plt.xticks(rotation=45) 
plt.show() 
```

```python
#4 
sns.boxplot(x="test preparation course", y="math score", data=df) 
plt.title("Подготовка к тесту и результаты по математике") 
plt.show() 
sns.boxplot(x="test preparation course", y="reading score", data=df) 
plt.title("Подготовка к тесту и результаты по чтению") 
plt.show() 
sns.boxplot(x="test preparation course", y="writing score", data=df) 
plt.title("Подготовка к тесту и результаты по письму") 
plt.show() 
```

```python
#5 
sns.heatmap(df[["math score", "reading score", "writing score"]].corr(), annot=True, cmap="Blues") 
plt.title("Корреляция между предметами") 
plt.show()
def visualize_students(df): 
    sns.pairplot(df[["math score", "reading score", "writing score"]]) 
    plt.suptitle("Парные зависимости между предметами", y=1.02) 
    plt.show() 
visualize_students(df)
```
